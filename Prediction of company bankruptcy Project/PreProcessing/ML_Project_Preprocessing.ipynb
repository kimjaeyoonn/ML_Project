# 필요한 라이브러리 설치

```python
# 필요한 라이브러리 설치
import copy
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.impute import KNNImputer
import warnings
warnings.filterwarnings('ignore')
```

# 데이터 불러오기

```python
# 파일 받아오기
df1 =pd.read_csv("1year.csv")
df2 =pd.read_csv("2year.csv")
df3 =pd.read_csv("3year.csv")
df4 =pd.read_csv("4year.csv")
df5 =pd.read_csv("5year.csv")
```

# ID 제거

```python
#ID 열은 제거
df1 = df1.drop(['id'], axis=1)
df2 = df2.drop(['id'], axis=1)
df3 = df3.drop(['id'], axis=1)
df4 = df4.drop(['id'], axis=1)
df5 = df5.drop(['id'], axis=1)
df_list = [df1, df2, df3, df4, df5]
```

# 중복 데이터 확인 및 제거

```python
for i in range(5):
    if df_list[i].duplicated().sum()>0:
        # 중복제거
        print(df_list[i]," : 중복 데이터 Check!")
        print(df_list[i]," : 중복된 데이터 수 : ",df_list[i].duplicated().sum())
        df_list[i].drop_duplicates(subset=None, keep='first', inplace=True, ignore_index=True)
        print("제거 후 :" ,df_list[i].duplicated().sum())
```

# 데이터 라벨링

```python
df1 = df1.replace({'bankruptcy':1},{'bankruptcy':5})
df1 = df1.replace({'bankruptcy':0},{'bankruptcy':-5})

df2 = df2.replace({'bankruptcy':1},{'bankruptcy':4})
df2 = df2.replace({'bankruptcy':0},{'bankruptcy':-4})

df3 = df3.replace({'bankruptcy':1},{'bankruptcy':3})
df3 = df3.replace({'bankruptcy':0},{'bankruptcy':-3})

df4 = df4.replace({'bankruptcy':1},{'bankruptcy':2})
df4 = df4.replace({'bankruptcy':0},{'bankruptcy':-2})

df5 = df5.replace({'bankruptcy':1},{'bankruptcy':1})
df5 = df5.replace({'bankruptcy':0},{'bankruptcy':-1})
```

# 데이터 합치기

```python
df = pd.concat([df1, df2, df3, df4, df5])
```

# SCALING

```python
# unseen도 
#스케일링 전처리 , feature 값과 Y값 분리 
df_dataset = df.drop(labels=['bankruptcy'], axis=1)
df_y = pd.DataFrame(df['bankruptcy'])
df_y.reset_index(inplace=True)
df_y.drop(['index'], axis=1, inplace=True)
col_names = df_dataset.columns

from sklearn.preprocessing import StandardScaler
scaler_std = StandardScaler()
df_standard = scaler_std.fit_transform(df_dataset)
df_dataset = pd.DataFrame(df_standard, columns=col_names)

df = pd.concat([df_dataset, df_y], axis=1)

# X , Y 나누기
df_dataset = df.drop(labels=['bankruptcy'], axis=1)
df_y = pd.DataFrame(df['bankruptcy'])
```

# 1. Iterative 결측치 대체 방법

```python
# Iterative Imputer를 이용한 다중대치
dummy = pd.get_dummies(df_dataset)
Mdf_dataset = pd.DataFrame(IterativeImputer(verbose=False).fit_transform(dummy)
                           ,columns = df_dataset.columns)
Mdf = pd.concat([Mdf_dataset, df_y], axis=1)
```

# 2. 단순 대치법( 열과 행 )

```python
# 상위 5개 지워
Sdf = df.drop(['attr37', 'attr27', 'attr60', 'attr45', 'attr21'], axis=1)
# 결측치 존재하는 rows에 대해서는 단순 대치
Sdf = Sdf.dropna(axis=0)
```

# 3. KNN 결측치 대체 방법

```python
imputer=KNNImputer(n_neighbors=5)

Kdf_dataset = imputer.fit_transform(df_dataset)
Kdf_dataset = pd.DataFrame(Kdf_dataset, columns = df_dataset.columns)
Kdf = pd.concat([Kdf_dataset, df_y], axis=1)
```

# 데이터 라벨링 1-> 1, 0 ->-1

# Unseen Data 추출

```python
Kdf1 = copy.deepcopy(Kdf)
Mdf1 = copy.deepcopy(Mdf)
Sdf1 = copy.deepcopy(Sdf)
def extrating_unseen(df):
    print(df['bankruptcy'].unique())
    label_nb = [-5,-4,-3,-2,-1]
    unseen = pd.DataFrame
    for i in label_nb:
        unseen_data = df[df['bankruptcy']==i].sample(n=2)
        for j in range(2):
            df = df.drop([unseen_data.index[j],], axis = 0) 
            if(unseen.empty):
                unseen = unseen_data
            else:
                unseen = pd.concat([unseen, unseen_data]) 
            
    label_b = [1,2,3,4,5]
    for i in label_b:
        unseen_data = df[df['bankruptcy']==i].sample(n=10)
        for j in range(10):
            df = df.drop([unseen_data.index[j],], axis = 0)
            unseen = pd.concat([unseen, unseen_data]) 
#     df.reset_index(inplace=True)
#     unseen.reset_index(inplace=True)
#     df = df.drop(['index'], axis = 1)
#     unseen = unseen.drop(['index'], axis = 1)
    return df, unseen

Kdf1, Kunseen = extrating_unseen(Kdf)
Mdf1, Munseen = extrating_unseen(Mdf)
Sdf1, Sunseen = extrating_unseen(Sdf)
```

# Each 학습 DF 와 Unseen 생성

```python
Kdf_names = ['Kdf_5year.csv','Kdf_4year.csv','Kdf_3year.csv','Kdf_2year.csv','Kdf_1year.csv']
Mdf_names = ['Mdf_5year.csv','Mdf_4year.csv','Mdf_3year.csv','Mdf_2year.csv','Mdf_1year.csv']
Sdf_names = ['Sdf_5year.csv','Sdf_4year.csv','Sdf_3year.csv','Sdf_2year.csv','Sdf_1year.csv']
def extrating_df(df, names):
    df_list = []
    for i in range(1,6):
        df_nb = df[df['bankruptcy']== -i]
        df_b = df[df['bankruptcy']== i]
        df_sum = pd.concat([df_nb, df_b])
        df_list.append(df_sum)
    for i in range(5):
        df_list[i].to_csv(names[i],index=False)    


extrating_df(Kdf1, Kdf_names)
extrating_df(Mdf1, Mdf_names)
extrating_df(Sdf1, Sdf_names)

Kunseen.to_csv('Kdf_unseen_10label.csv',index=False)
Munseen.to_csv('Mdf_unseen_10label.csv',index=False)
Sunseen.to_csv('Sdf_unseen_10label.csv',index=False)

Kdf1.to_csv('Kdf_10label.csv',index=False)
Mdf1.to_csv('Mdf_10label.csv',index=False)
Sdf1.to_csv('Sdf_10label.csv',index=False)
```

```python
def relabeling(df):
    df = df.replace({'bankruptcy':-5},{'bankruptcy':0})
    df = df.replace({'bankruptcy':-4},{'bankruptcy':0})
    df = df.replace({'bankruptcy':-3},{'bankruptcy':0})
    df = df.replace({'bankruptcy':-2},{'bankruptcy':0})
    df = df.replace({'bankruptcy':-1},{'bankruptcy':0})
    return df

Kdf_re = relabeling(Kdf1)
Mdf_re = relabeling(Mdf1)
Sdf_re = relabeling(Sdf1)
dfre_list = [Kdf_re,Mdf_re,Sdf_re]

Kunseen_re = relabeling(Kunseen)
Munseen_re = relabeling(Munseen)
Sunseen_re = relabeling(Sunseen)
unseenre_list = [Kunseen_re,Munseen_re,Sunseen_re]

dfre_names = ['Kdf_6label.csv','Mdf_6label.csv','Sdf_6label.csv']
unseenre_names = ['Kdf_unseen_6label.csv','Mdf_unseen_6label.csv','Sdf_unseen_6label.csv']
def tocsv(relist, names):
    for i in range(3):
        relist[i].to_csv(names[i],index=False) 
        
tocsv(dfre_list, dfre_names)
tocsv(unseenre_list,unseenre_names)
```
